## Business Understanding

Our task for this lab assignment is to predict the outcome of a League of Legends match by analysing the statistics and data collected from previous matches. 
League of Legends is one of the largest gaming e-sports played on a collegiate and professional level. Competitions are organized by Riot Games in North America and Europe. The League Championship Series (LCS) is located in Berlin and Los Angeles which consists of 10 professional teams in each continent. These competitions have a large number of players and tournaments with a huge cash prize. Being able to predict the outcome of the matches by analysing the outcomes of previous matches would be a great advantage to the players. 
Players could use the data from past competitions to know which game strategies lead to wins and which strategies lead to match losses. Using the data, players will know what items to buy, how many kills are required, what turrets and inhibitors to use in order to win the game. This would help the teams to develop a strategy to win the match and, thus, win the cash prize.

### Choosing the Evaluation Metric

We are concerned with the wins and losses of a player in a particular match. That means, our model should be able to predict the two outcomes as accurately as possible. Considering wins as the positive and losses as the negative outcome, we want to maximize the True positives and True negatives in order to achieve the correct prediction. Furthermore, an incorrect prediction is not  very critical in our business case, that's why we need not worry much about the False positives and False negatives.

Therefore, using **Accuracy score** would be the best evaluation metric for our dataset which also satisfies the purpose of our business case.

$$Accuracy = \frac{TP+TN}{TP+TN+FP+FN}$$

## 2. Modeling

The following sections show 3 types of Wide and Deep networks. This will be done by using different set of cross-columns and changing the number of layers in the deep branch. Let us see which one is the best.

```python
# Model 1 : 3 Deep branches
cross_columns = [['kills', 'deaths', 'assists'],
                 ['championid', 'champlvl' ]]

embed_branches = []
X_ints_train = []
X_ints_test = []
all_inputs = []
all_branch_outputs = []

for cols in cross_columns:
    # encode crossed columns as ints for the embedding
    enc = LabelEncoder()
    
    # create crossed labels
    X_crossed_train = df_train[cols].apply(lambda x: '_'.join(str(x)), axis=1)
    X_crossed_test = df_test[cols].apply(lambda x: '_'.join(str(x)), axis=1)
    
    enc.fit(np.hstack((X_crossed_train.values,  X_crossed_test.values)))
    X_crossed_train = enc.transform(X_crossed_train)
    X_crossed_test = enc.transform(X_crossed_test)
    X_ints_train.append( X_crossed_train )
    X_ints_test.append( X_crossed_test )
    
    # get the number of categories
    N = max(X_ints_train[-1]+1) # same as the max(df_train[col])
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name = '_'.join(cols))
    all_inputs.append(inputs)
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name = '_'.join(cols)+'_embed')(inputs)
    x = Flatten()(x)
    all_branch_outputs.append(x)
    
# merge the branches together
wide_branch = concatenate(all_branch_outputs, name='wide_concat')

# reset this input branch
all_branch_outputs = []
# add in the embeddings
for col in categorical_headers_ints:
    # encode as ints for the embedding
    X_ints_train.append( df_train[col].values )
    X_ints_test.append( df_test[col].values )
    
    # get the number of categories
    N = max(X_ints_train[-1]+1) # same as the max(df_train[col])
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name=col)
    all_inputs.append(inputs)
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name=col+'_embed')(inputs)
    x = Flatten()(x)
    all_branch_outputs.append(x)
    
# also get a dense branch of the numeric features
all_inputs.append(Input(shape=(X_train_num.shape[1],),
                        sparse=False,
                        name='numeric_data'))

x = Dense(units=20, activation='relu',name='numeric_1')(all_inputs[-1])
all_branch_outputs.append( x )

# merge the branches together
deep_branch = concatenate(all_branch_outputs,name='concat_embeds')
deep_branch = Dense(units=50,activation='relu', name='deep1')(deep_branch)
deep_branch = Dense(units=25,activation='relu', name='deep2')(deep_branch)
deep_branch = Dense(units=10,activation='relu', name='deep3')(deep_branch)
    
final_branch = concatenate([wide_branch, deep_branch],name='concat_deep_wide')
final_branch = Dense(units=1,activation='sigmoid',name='combined')(final_branch)

model1 = Model(inputs=all_inputs, outputs=final_branch)
```

```python
SVG(model_to_dot(model1).create(prog='dot', format='svg'))
```

```python
%%time

model1.compile(optimizer='adagrad',
              loss='mean_squared_error',
              metrics=['accuracy'])
# lets also add the history variable to see how we are doing
model_history_1 = model1.fit(X_ints_train+ [X_train_num],
                    y_train, 
                    epochs=5, 
                    batch_size=32, 
                    verbose=1, 
                    validation_data = (X_ints_test + [X_test_num], y_test))
```

```python
yhat = np.round(model1.predict(X_ints_test + [X_test_num]))
model_acc_1 = mt.accuracy_score(y_test,yhat)
print(mt.confusion_matrix(y_test,yhat), model_acc_1)
```

```python
# Model 2 : 5 Deep branches
cross_columns = [['magicdmgtochamp', 'truedmgtochamp', 'physdmgtochamp'],
                 ['championid', 'champlvl'],
                 ['doublekills', 'triplekills', 'quadrakills', 'pentakills']]

embed_branches = []
X_ints_train = []
X_ints_test = []
all_inputs = []
all_branch_outputs = []

for cols in cross_columns:
    # encode crossed columns as ints for the embedding
    enc = LabelEncoder()
    
    # create crossed labels
    X_crossed_train = df_train[cols].apply(lambda x: '_'.join(str(x)), axis=1)
    X_crossed_test = df_test[cols].apply(lambda x: '_'.join(str(x)), axis=1)
    
    enc.fit(np.hstack((X_crossed_train.values,  X_crossed_test.values)))
    X_crossed_train = enc.transform(X_crossed_train)
    X_crossed_test = enc.transform(X_crossed_test)
    X_ints_train.append( X_crossed_train )
    X_ints_test.append( X_crossed_test )
    
    # get the number of categories
    N = max(X_ints_train[-1]+1) # same as the max(df_train[col])
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name = '_'.join(cols))
    all_inputs.append(inputs)
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name = '_'.join(cols)+'_embed')(inputs)
    x = Flatten()(x)
    all_branch_outputs.append(x)
    
# merge the branches together
wide_branch = concatenate(all_branch_outputs, name='wide_concat')

# reset this input branch
all_branch_outputs = []
# add in the embeddings
for col in categorical_headers_ints:
    # encode as ints for the embedding
    X_ints_train.append( df_train[col].values )
    X_ints_test.append( df_test[col].values )
    
    # get the number of categories
    N = max(X_ints_train[-1]+1) # same as the max(df_train[col])
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name=col)
    all_inputs.append(inputs)
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name=col+'_embed')(inputs)
    x = Flatten()(x)
    all_branch_outputs.append(x)
    
# also get a dense branch of the numeric features
all_inputs.append(Input(shape=(X_train_num.shape[1],),
                        sparse=False,
                        name='numeric_data'))

x = Dense(units=20, activation='relu',name='numeric_1')(all_inputs[-1])
all_branch_outputs.append( x )

# merge the branches together
deep_branch = concatenate(all_branch_outputs,name='concat_embeds')
deep_branch = Dense(units=150,activation='relu', name='deep1')(deep_branch)
deep_branch = Dropout(0.20)(deep_branch)
deep_branch = Dense(units=100,activation='relu', name='deep2')(deep_branch)
deep_branch = Dense(units=50,activation='relu', name='deep3')(deep_branch)
deep_branch = Dropout(0.30)(deep_branch)
deep_branch = Dense(units=20,activation='relu', name='deep4')(deep_branch)
deep_branch = Dropout(0.50)(deep_branch)
deep_branch = Dense(units=10,activation='relu', name='deep5')(deep_branch)
    
final_branch = concatenate([wide_branch, deep_branch],name='concat_deep_wide')
final_branch = Dense(units=1,activation='sigmoid',name='combined')(final_branch)

model2 = Model(inputs=all_inputs, outputs=final_branch)
```

Let us compare the performance of our best model with a standard Multi-layer perceptron using ROC and area under the curve.

```python
def compute_roc(probas, y_test, n_classes):
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test, probas)
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Compute macro-average ROC curve and ROC area
    # First aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[i], tpr[i])

    # Finally average it and compute AUC
    mean_tpr /= n_classes

    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    
    return fpr, tpr, roc_auc 
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100, ), 
                    activation='relu', # type of non-linearity, every layer
                    solver='adam', 
                    alpha=1e-4, # L2 penalty
                    batch_size= 'auto', # min of 200, num_samples
                    learning_rate='constant', # adapt learning? only for sgd
                    learning_rate_init=0.1, # only SGD
                    power_t=0.0,    # only SGD with inverse scaling of learning rate
                    max_iter=75, # stopping criteria
                    shuffle=True, 
                    random_state=1, 
                    tol=0, # for stopping
                    verbose=False, 
                    warm_start=False, 
                    momentum=0.9, # only SGD
                    nesterovs_momentum=False, # only SGD
                    early_stopping=False, 
                    validation_fraction=0.0, # only if early_stop is true
                    beta_1=0.9, # adam decay rate of moment
                    beta_2=0.999, # adam decay rate of moment
                    epsilon=1e-08) # adam numerical stabilizer

clf.fit(X_train,y_train)
```

```python
y_score = np.round(model1.predict(X_ints_test + [X_test_num]))
yhat = clf.predict(X_test)
error_rate_mlp = 1-mt.accuracy_score(y_test,yhat)

fpr, tpr, roc_auc = compute_roc(y_score, y_test, n_classes)
fpr_mlp, tpr_mlp, roc_auc_mlp = compute_roc(yhat, y_test, n_classes)
```

```python
plt.figure(figsize=(8, 5))

plt.plot(fpr["macro"], tpr["macro"], label='macro-average ROC curve for W-D network (area = {0:0.2f})'.format(roc_auc["macro"]),
         color='navy', linestyle='--', linewidth=2)
plt.plot(fpr_mlp["macro"], tpr_mlp["macro"], label='macro-average ROC curve for MLP network(area = {0:0.2f})'.format(roc_auc_mlp["macro"]),
         color='red', linestyle='--', linewidth=2)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

As we can see from the ROC curve for both the models, the MLP does slightly better with the True positive rate in the False positive rate interval of 0.0 - 0.2. Further to that point, both of them appear to do equally well as we can see them converging.

Now, let us compare the two models and see the statistical difference in their performances. To do this, we will use Binomial distribution to calculate the standard deviation in the error rates of the two models. For this comparison, we will use a confidence interval of 95%.

```python
error_rate = 1 - model_acc_1

Z = 1.96 # at 95 % confidence
error_diff = np.abs(error_rate - error_rate_mlp)
std_dev = ((error_rate * (1-error_rate)) + (error_rate_mlp * (1-error_rate_mlp))) / y_test.shape[0]

print("Range of:", error_diff - (Z * np.sqrt(std_dev)), error_diff + (Z * np.sqrt(std_dev)))
```

Okay, so the difference lies in the range of 0.53% - 3.04%, which means that the MLP network is better than our implemented model of wide and deep network by the given range of difference at 95% confidence interval.
